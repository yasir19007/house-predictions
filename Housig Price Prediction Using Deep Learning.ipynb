# **Train Test Split**

```python
X = df.drop('price', axis=1)
y= df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)

# X = scaler.fit_transform(X)
# X

X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
#  the shape of the input data is 19 inputs or features thats why we have taken 19
#  neurons in the first layer

model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
# validation data is used to see the LOSE using test data after each epoch during training on the training data
# Validation data : After each epoch of training on the training data, the model will run the test data to check the loss on our data

model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=132, epochs=400)
```

### **Comparing Training Loss vs Validation Loss**

```python
losses = pd.DataFrame(model.history.history)

losses
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
model_prediction = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, model_prediction))
```

```python
mean_absolute_error(y_test, model_prediction)
```
